```python
import pandas as pd
url = 'https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/'
data = pd.read_html(url, header=None)
df = data[0]
df.columns = df.iloc[0]  # ustaw pierwszy wiersz jako nazwy kolumn
df = df[1:]              # usuń ten wiersz z danych
df.reset_index(drop=True, inplace=True)

#1. Zmiana nazw kolumn
df = df.rename(columns={
    'POS': 'POZYCJA',
    'TITLE': 'TYTUŁ',
    'ARTIST': 'ARTYSTA',
    'YEAR': 'ROK',
    'HIGH POSN': 'NAJWYŻSZA POZYCJA'
})

df['ARTYSTA'] = df['ARTYSTA'].str.upper().str.strip()
print("Nowe nazwy kolumn:")
print(df.columns)

#2. Ilość pojedynczych artystów
unikalni_artysci = df['ARTYSTA'].nunique()
print(f"Ilość pojedynczych artystów na liście: {unikalni_artysci}")
print("Lista pojedynczych artystów:")
print(df['ARTYSTA'].unique())

#3.Najczęściej pojawiające się zespoły
najczestsze = df['ARTYSTA'].value_counts()
print("Top 10 artystów z największą liczbą albumów na liście:")
print(najczestsze.head(10))

#4.Zmiana nagłówków
df.columns = [col.capitalize() for col in df.columns]
print("Nowe nagłówki kolumn:")
print(df.columns)

#5. Usuwanie kolumny
if 'Najwyższa pozycja' in df.columns:
    df.drop(columns=['Najwyższa pozycja'], inplace=True)
    print("Kolumny po usunięciu:")
    print(df.columns)

#6.Najwięcej albumów
roki_statystyka = df['Rok'].value_counts().sort_values(ascending=False)
najpopularniejszy_rok = roki_statystyka.idxmax()
liczba_albumow = roki_statystyka.max()
print(f"Najwięcej albumów na liście pochodzi z roku {najpopularniejszy_rok}")

#7. Wydane albumy między 1960 - 1990
df['Rok'] = pd.to_numeric(df['Rok'], errors='coerce')
albumy_1960_1990 = df[(df['Rok'] >= 1960) & (df['Rok'] <= 1990)]
print(f"Wydane albumy między 1960–1990: {len(albumy_1960_1990)}")

#8. Najmłodszy album
df['Rok'] = pd.to_numeric(df['Rok'], errors='coerce')
najmlodszy_rok = df['Rok'].max()
print(f"Najmłodszy album to: {najmlodszy_rok}")

#9. Najwcześniej wydane albumy każdego artysty
df['Rok'] = pd.to_numeric(df['Rok'], errors='coerce')
# Znalezienie najwcześniejszego albumu dla każdego artysty
najwczesniejsze_albumy = df.loc[df.groupby('Artysta')['Rok'].idxmin()][['Artysta', 'Tytuł', 'Rok']]
najwczesniejsze_albumy = najwczesniejsze_albumy.sort_values(by='Rok')
print("Najwcześniej wydane albumy każdego artysty:")
print(najwczesniejsze_albumy)

#10.Zapi listy jako plik CSV
najwczesniejsze_albumy.to_csv('najwczesniejsze_albumy.csv', index=False, encoding='utf-8-sig')
print("Zapisano plik: najwczesniejsze_albumy.csv")
```
